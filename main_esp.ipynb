## 6. Escribe una funci贸n que devuelva el factorial de un n煤mero.

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        result = 1
    for n in range(1, n+1):
        result = n * result
    return result
```

## 11. Escribe una funci贸n que reciba una cadena de palabras separadas por comas y devuelva una cadena de palabras separadas por comas ordenadas alfab茅ticamente.

`NOTA: Puedes usar sorted pero no split y definitivamente no join! お`

```python
def sort_alpha(string):
    # Convertimos el string en una lista de palabras separadas por comas
    palabras = []
    palabra = ''
    for caracter in string:
        if caracter != ',':
            palabra += caracter
        else:
            palabras.append(palabra)
            palabra = ''
    palabras.append(palabra)  
    
    # Ordenamos la lista de palabras
    palabras_ordenadas = sorted(palabras)
    
    # Creamos una cadena con las palabras ordenadas separadas por comas
    cadena_ordenada = ''
    for index, palabra in enumerate(palabras_ordenadas):
        if index != len(palabras_ordenadas) - 1:
            cadena_ordenada += palabra + ','
        else:
            cadena_ordenada += palabra
    return cadena_ordenada
```

```python
# Para testear tu funci贸n 
test_alpha(sort_alpha)
```

## 12. Escribe una funci贸n para verificar si una contrase帽a dada es fuerte (al menos 8 caracteres, al menos una min煤scula, al menos una may煤scula, al menos un n煤mero y al menos un car谩cter especial). Deber铆a devolver True si es fuerte y False si no lo es.

```python
import re

def check_pass(password):
    #verifica la logitud
    if len(password) < 8:
        return False
    
    # Verificamos si hay al menos una min煤scula
    if not re.search("[a-z]", password):
        return False
    
    # Verificamos si hay al menos una may煤scula
    if not re.search("[A-Z]", password):
        return False
    
    # Verificamos si hay al menos un n煤mero
    if not re.search("[0-9]", password):
        return False
    
    # Verificamos si hay al menos un car谩cter especial
    if not re.search("[#@!$%&()^*{}\[\]]+", password):
        return False
    
    # Si la contrase帽a pasa todas las verificaciones, es fuerte
    return True
```

```python
# Para testear tu funci贸n
test_pass(check_pass)
```
